```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
def numerical_analysis(dataframe, col_name,cat_col=None,bins= "auto"):
    #create the figure
    fig= plt.figure(figsize=(15,10))
    grid = GridSpec(nrows=2,ncols=2,figure=fig)
    ax1= fig.add_subplot(grid[0,1])
    ax2= fig.add_subplot(grid[0,0])
    ax3= fig.add_subplot(grid[1,:])

    #plot the kde plot
    sns.kdeplot(data=dataframe, x=col_name, hue=cat_col,ax=ax1)

    #plot the boxplot
    sns.boxplot(data=dataframe, x=cat_col, y=col_name, ax=ax2)

    #plot the histogram
    sns.histplot(data=dataframe, x=col_name,hue=cat_col, bins=bins, ax=ax3, kde=True)
    plt.tight_layout()
    plt.show()
```

```python
#this is to check normality, weather target is following normal distribution or not

from scipy import stats as stats_plots
import matplotlib.pyplot as plt
stats_plots.probplot(df['Lifespan'], dist="norm", plot=plt)
plt.show()


#Inference
# The Q-Q plot indicates that the Lifespan data closely follows a normal distribution, as most points align along the reference line.   
```

```python
from scipy.stats import jarque_bera
def test_for_normality(data, col_name, alpha=0.05):
    data = data[col_name]
    print("jarque bera test for normality")
    _,p_val = jarque_bera(data)
    print(f'P-value: {p_val}')

    if p_val <= alpha:
        print(f'reject the null hypothesis. {col_name} is normally distributed.')
    else:
        print(f'Fail to Reject the null hypothesis. {col_name} is not normally distributed.')

test_for_normality(df,'Lifespan')
```

```python
def multivariate_analysis(dataframe, num_col_1, num_col_2, cat_col_1=None):
    #create the figure
    fig, (ax1,ax2) = plt.subplots(2,2,figsize=(15,7.5))
    sns.lineplot(data=dataframe,x=num_col_1,y=num_col_2,hue = cat_col_1,ax=ax1[0])

    sns.catplot(data=dataframe,x=num_col_1,y=num_col_2,hue = cat_col_1,ax=ax1[1])

    sns.relplot(data=dataframe,x=num_col_1,y=num_col_2,hue = cat_col_1,ax=ax2[0])

    sns.scatterplot(data=dataframe,x=num_col_1,y=num_col_2,hue = cat_col_1,ax=ax2[1])

    plt.tight_layout()
    plt.show()
```

```python
multivariate_analysis(df,'Sleep','Lifespan')
```

```python
from scipy.stats import chi2_contingency
def chi_2_test(dataframe,col1,col2,alpha =0.5):
    data = (dataframe.loc[:,[col1,col2]])

    contigency_table = pd.crosstab(data[col1],data[col2])

    _,p_val,_,_ = chi2_contingency(contigency_table)
    print(f'P-value: {p_val}')
    if p_val<=alpha:
        print(f'reject the null hypothesis. {col1} and {col2} are independent.')
    else:
        print(f'Fail to Reject the null hypothesis. {col1} and {col2} are dependent.')
```

```python
def change_column_names(data:pd.DataFrame):
    return (
        data.rename(str.lower,axis=1)
        .rename({
            'occupation_type':'Occupation',
            'avg_work_hours_per_day':'Work',
            'avg_rest_hours_per_day':'Rest',
            'avg_sleep_hours_per_day':'Sleep',
            'avg_exercise_hours_per_day':'Exercise',
            'age_at_death':'Lifespan'},axis=1)
    ).drop('id',axis=1)
```

```python
def perform_data_cleaning(data:pd.DataFrame, saved_data_path="cleaned_data.csv"):
    cleaned_data = (
        data
        .pipe(change_column_names)
    )

    #save the data
    cleaned_data.to_csv(saved_data_path, index=False)
```
